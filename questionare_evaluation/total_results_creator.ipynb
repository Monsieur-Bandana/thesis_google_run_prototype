# 1st Iteration

```python
import pandas as pd
import os
import re

csv_files = [file for file in os.listdir('.') if file.endswith('1.csv')]

total_df = pd.DataFrame()
for file in csv_files:
    df = pd.read_csv(file,sep=';', index_col=False)

    total_df = pd.concat([total_df, df], axis=1,join="outer")


total_df = total_df.transpose()
total_df = total_df[::-1]
total_df
```
